```python
df = pd.read_csv("../data/raw/creditcard.csv")
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, average_precision_score
import numpy as np

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

f1_scores = []
aucpr_scores = []

for train_idx, val_idx in skf.split(X, y):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = LogisticRegression(max_iter=1000, class_weight="balanced")
    model.fit(X_train, y_train)

    preds = model.predict(X_val)
    probs = model.predict_proba(X_val)[:, 1]

    f1_scores.append(f1_score(y_val, preds))
    aucpr_scores.append(average_precision_score(y_val, probs))

print("F1 Mean:", np.mean(f1_scores))
print("F1 Std:", np.std(f1_scores))
print("AUC-PR Mean:", np.mean(aucpr_scores))
print("AUC-PR Std:", np.std(aucpr_scores))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, average_precision_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Logistic Regression
lr = LogisticRegression(max_iter=1000, class_weight="balanced")
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_prob = lr.predict_proba(X_test)[:,1]

# XGBoost
xgb = XGBClassifier(
    n_estimators=200,
    max_depth=4,
    learning_rate=0.1,
    scale_pos_weight=len(y_train[y_train==0])/len(y_train[y_train==1]),
    eval_metric="logloss"
)

xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
xgb_prob = xgb.predict_proba(X_test)[:,1]

results = pd.DataFrame({
    "Model": ["Logistic Regression", "XGBoost"],
    "F1": [f1_score(y_test, lr_pred), f1_score(y_test, xgb_pred)],
    "AUC-PR": [average_precision_score(y_test, lr_prob),
               average_precision_score(y_test, xgb_prob)]
})

results
```

XGBoost achieved higher F1-score and AUC-PR compared to Logistic Regression, indicating better performance in detecting fraud cases. However, Logistic Regression remains valuable for interpretability. Therefore, XGBoost is selected as the final model due to its superior predictive capability on imbalanced data.
